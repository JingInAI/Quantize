```python
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

loc = np.arange(4)
points = np.meshgrid(loc, loc, loc)

for x, y, z in zip(*points):
    ax.scatter(x, y, z, c='k', marker='o', s=50)

plt.axis('off')
plt.show()
fig.savefig('logo.png', dpi=300, bbox_inches='tight', pad_inches=0, transparent=True)
```
